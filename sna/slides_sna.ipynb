```python
G = nx.Graph()
G.add_edge('Mary', 'Bob')
G.add_edge('Mary', 'Alice')
G.add_edge('Edgar', 'Alice')
G.add_edge('Edgar', 'Julia')
G.add_edge('Alice', 'Bob')
G.add_edge('Alice', 'Julia')
G.add_edge('Alice', 'Rafael')
G.add_edge('Julia', 'Jim')
G.add_edge('Bob', 'Rafael')
G.add_edge('Jim', 'Rafael')
G.add_edge('Rafael', 'Aldo')
G.add_edge('Aldo', 'Stefan')
G.add_edge('Aldo', 'Pierre')
```

```python
from networkx.algorithms.centrality import closeness_centrality, eigenvector_centrality, betweenness_centrality, degree_centrality
import operator

G.number_of_nodes()
G.number_of_edges()

print('Degree:')
deg = degree_centrality(G)
sorted(deg.items(), key=lambda x: -x[1])

print('Closeness:')
close = closeness_centrality(G)
sorted(close.items(), key=lambda x: -x[1])

print('Eigenvector/Pagerank:')
eig = eigenvector_centrality(G)
sorted(eig.items(), key=lambda x: -x[1])

print('Betweeness:')
betw = betweenness_centrality(G)
sorted(betw.items(), key=lambda x: -x[1])
```

```python
options = {
     'node_color': 'lightgray',
     'width': 3,
     'font_weight': 'normal',
     'with_labels': True,
     'alpha': 0.98,
}

seed = 154
numpy.random.seed(seed)
nx.draw_spring(G, **options, node_size=1500)
plt.savefig('out/sna_normal.png', dpi=300)
```
