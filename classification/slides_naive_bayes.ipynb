```python
import math
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
np.random.seed(seed=42)
age1 = np.random.normal(loc=25, scale=5, size=9)
age2 = np.random.normal(loc=35, scale=3, size=9)

age1
age2
```

```python
mean1 = age1.mean()
std1 = age1.std()
mean1
std1

mean2 = age2.mean()
std2 = age2.std()
mean2
std2
```

```python
v = 36

stats.norm(mean1, std1).pdf(v)
stats.norm(mean2, std2).pdf(v)
```

```python
fig, ax = plt.subplots(figsize=(9, 6));  

ax.grid(True, which='both');
#ax.set_aspect('auto')
#ax.set_xlim(xlim);
#ax.set_ylim(ylim);
ax.tick_params(axis='both', which='major', labelsize=20);


x1 = np.linspace(mean1 - 4*std1, mean1 + 4*std1, 100)
plt.plot(x1, stats.norm.pdf(x1, mean1, std1), label="paid")
#plt.show()

x2 = np.linspace(mean2 - 4*std2, mean2 + 4*std2, 100)
plt.plot(x2, stats.norm.pdf(x2, mean2, std2), label="default")

plt.legend(prop={'size': 20});

plt.ylabel('Probability Density', fontsize=26)

#ax.axvline(x=v, color='k');

plt.savefig('out/pdfs.png', bbox_inches='tight');
plt.show();
```
