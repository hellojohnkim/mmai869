```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
from scipy.special import expit, logit, softmax

def relu(x):
    return np.maximum(x, 0)

def leaky_relu(x):
    return np.where(x > 0, x, x * 0.1)  


def swish(x):
    return x * expit(x)

def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
def show_plot(x, y, name, xlim=[-5, 5], ylim=[-5, 5]):
    fig, ax = plt.subplots(figsize=(9, 6));  

    ax.axhline(y=0, color='k');
    ax.axvline(x=0, color='k');
    ax.grid(True, which='both');
    #ax.set_aspect('auto')
    ax.set_xlim(xlim);
    ax.set_ylim(ylim);
    ax.tick_params(axis='both', which='major', labelsize=20);
    
    t = ax.text(0.5, 0.99, name, horizontalalignment='center',
        verticalalignment='center',
        fontsize=30, fontweight='normal', color='tab:blue',
        transform=ax.transAxes);

    t.set_bbox(dict(facecolor='white', alpha=0.8, edgecolor="none", pad=0));
    
    ax.plot(x, y, linewidth=5);

    plt.savefig('out/'+name+'.png');
    plt.show();
    
```

```python
x = np.linspace(-10.0, 10.0, 100)
print(x)
```

```python
show_plot(x, x, 'identity')
```

```python
show_plot(x, expit(x), 'sigmoid', ylim=[-0.1, 1.1])
```

```python
show_plot(x, softplus(x), 'softplus')
```

```python
show_plot(x, relu(x), 'relu')
```

```python
show_plot(x, leaky_relu(x), 'leaky_relu')
```

```python
show_plot(x, swish(x), 'swish')
```
